## 4.2 Boston House Price Prediction

In the next questions we are going to work using the dataset *Boston*. This dataset measures the influence of socioeconomical factors on the price of several estates of the city of Boston. This dataset has 506 instances, each one characterized by 13 features:

* CRIM - per capita crime rate by town
* ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
* INDUS - proportion of non-retail business acres per town.
* CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
* NOX - nitric oxides concentration (parts per 10 million)
* RM - average number of rooms per dwelling
* AGE - proportion of owner-occupied units built prior to 1940
* DIS - weighted distances to five Boston employment centres
* RAD - index of accessibility to radial highways
* TAX - full-value property-tax rate per 10,000 USD
* PTRATIO - pupil-teacher ratio by town
* B - $1000(Bk - 0.63)^2$ where $Bk$ is the proportion of blacks by town
* LSTAT - % lower status of the population

Output variable:
* MEDV - Median value of owner-occupied homes in 1000's USD

**Note:** In this exercise we are going to predict the price of each estate, which is represented in the `MEDV` variable. It is important to remember that we are always aiming to predict `MEDV`, no matter which explanatory variables we are using. That means, in some cases we will use a subset of the 13 previously mentioned variables, while in other cases we will use all the 13 variables. # <center>SOLUCIÓN 4.1 KERAS Little Red Riding Hood Network</center> 

````python
import numpy as np
import pylab as pl
import keras
import pandas as pd

from keras.models import Sequential
from keras.layers import Dense, Activation
from sklearn.datasets.samples_generator import make_moons
````

### <font color='red'>ENTRADA Y SALIDA

````python
X = np.matrix([[1,1,0,0], [0,1,1,0],[0,0,0,1]])
print(X)

Y = np.matrix([[1,0,0,0], [0,0,1,1],[0,1,1,0]])
print(Y)
````

### <font color='red'> DEFINICIÓN DE MODELO

````python
# create model
model = Sequential()
model.add(Dense(units=3, input_dim=4, activation='relu'))
model.add(Dense(4, activation='sigmoid'))
````

### <font color='red'> ESTRUCTURA DE LA RED

````python
print(model.summary())
````

### <font color='red'> COMPILACIÓN DEL MODELO

````python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
````

````python
from keras.optimizers import SGD
model.compile(loss = 'binary_crossentropy',
              optimizer = SGD(lr=0.01, momentum=0.9, decay=0.0, nesterov=False),
              metrics = ['accuracy'])
````

###  <font color='red'>ESTIMACIÓN DE PARÁMETROS

````python
model.fit(X, Y, epochs=500, verbose=2)
````

### <font color='red'> PREDICCIÓN

````python
def lrrh_input(vals):
    return pd.DataFrame(vals,index=['big ears', 'big teeth', 'handsome', 'wrinkled'], columns=['input'])
````

````python
in_vals = [1, 1, 0, 0]
lrrh_input(in_vals)
````

````python
pred = model.predict(np.array([[1, 1,0,0]]))
print(pred)
````

````python
#Definición de output
df = pd.DataFrame(pred,columns=['scream', 'hug', 'offer food', 'kiss cheek'])
df = df.transpose()
df.columns = ['Output']
df
````

````python
in_vals = [0, 1, 1, 0]
lrrh_input(in_vals)
````

````python
df = pd.DataFrame(pred,columns=['scream', 'hug', 'offer food', 'kiss cheek'])
df = df.transpose()
df.columns = ['Output']
df
````

#### <font color='red'> 2. For each subset, generate a training partition and a test partition using a ratio of  70%−30%

````python
# train/test split
from sklearn.model_selection import train_test_split

Xd1 = pd.DataFrame(D1,columns=['RM'])
yd1 = D1['MEDV']

Xd2 = pd.DataFrame(D2,columns=['LSTAT'])
yd2 = D2['MEDV']

D1_Xtrain, D1_Xtest, D1_Ytrain, D1_Ytest = train_test_split(Xd1,yd1,test_size=0.30, random_state=42)
D2_Xtrain, D2_Xtest, D2_Ytrain, D2_Ytest = train_test_split(Xd2,yd2,test_size=0.30, random_state=44)
````

### <font color='red'><center>SET D1

````python
# scatter plot
plt.scatter(D1_Xtrain, D1_Ytrain,c='b',)
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.title('D1 TRAIN')
````

````python
# import
from sklearn.linear_model import LinearRegression
````

````python
# Initialize
linregD1 = LinearRegression(fit_intercept=True)
linregD1
````

````python
# Fit
linregD1.fit(D1_Xtrain,D1_Ytrain)
linregD1.coef_
````

````python
#Predicción
predD1= linregD1.predict(D1_Xtest)
predD1
````

````python
# scatter plot
plt.scatter(D1_Xtrain, D1_Ytrain,c='b',)
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.title('D1 TRAIN')

#Representación gráfica de la regresión
plt.plot(D1_Xtest, predD1, color='red', linewidth=3)
plt.show()
````

````python
#Evaluación a través del MSE modelo múltiple
from sklearn import metrics
print('MSE:', metrics.mean_squared_error(D1_Ytest, predD1))
````
